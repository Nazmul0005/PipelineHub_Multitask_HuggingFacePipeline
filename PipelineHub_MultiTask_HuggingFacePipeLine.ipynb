**Use Pretrained Model to Translation**

```python
from transformers import pipeline

# Define a function for translation with dynamic language selection
def translate_text(text, source_lang="en", target_lang="es"):
    # Load the translation pipeline dynamically based on source and target languages
    model_name = f"Helsinki-NLP/opus-mt-{source_lang}-{target_lang}"
    translator = pipeline("translation", model=model_name)

    # Translate the text
    result = translator(text, max_length=100)
    return result[0]["translation_text"]

# Example usage
source_language = "en"  # Source language: English
target_language = "es"  # Target language: Spanish
text = "Hello, how are you?"

translated_text = translate_text(text, source_lang=source_language, target_lang=target_language)
print(f"Translated Text ({source_language} to {target_language}): {translated_text}")
```

```python
from transformers import pipeline

def translate_text(source_language, target_language, text):
    # Define a mapping of source and target language combinations to model names
    model_mapping = {
        ("en", "fr"): "Helsinki-NLP/opus-mt-en-fr",
        ("en", "es"): "Helsinki-NLP/opus-mt-en-es",
        ("en", "bn"): "Helsinki-NLP/opus-mt-en-bn",
        ("fr", "en"): "Helsinki-NLP/opus-mt-fr-en",
        ("es", "en"): "Helsinki-NLP/opus-mt-es-en",
        ("bn", "en"): "Helsinki-NLP/opus-mt-bn-en",
        # Add more language pairs here as needed
    }

    # Check if the source-target pair is supported
    model_key = (source_language, target_language)
    if model_key not in model_mapping:
        raise ValueError(f"Translation from {source_language} to {target_language} is not supported.")

    # Load the corresponding pre-trained translation model
    model_name = model_mapping[model_key]
    translator = pipeline("text2text-generation", model=model_name)

    # Perform the translation
    result = translator(text)
    return result

# Example usage
source_language = "en"  # English
target_language = "es"  # Spanish
text = "I love programming."

# Translate the text
translated_text = translate_text(source_language, target_language, text)
print(translated_text)
```
